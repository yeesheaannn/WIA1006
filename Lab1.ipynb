# **2. Data Preprocessing**

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
total_rows = df.shape[0]
nan_rows = df.isnull().sum(axis=1).astype(bool).sum()
print('Number of records in total: {}\nNumber of records that have NaN values: {}'.format(total_rows, nan_rows))

# Handle missing values
df = df.dropna()
print('Number of records after removing NaN values: {}'.format(df.shape[0]))
```

```python
# Handle categorical variables
# List of features/variables to map
var =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the df list
df[var] = df[var].apply(binary_map)
```

```python
# Check the df dataframe now
df.head()
```

```python
# Handle the feature that has more than two status 
status = pd.get_dummies(df['furnishingstatus'])
```

```python
status.head()
```

```python
# Drop the first column from status df using 'drop_first = True'
status = pd.get_dummies(df['furnishingstatus'], drop_first = True)
```

```python
# Add the results to the original df dataframe
df = pd.concat([df, status], axis = 1)
```

```python
df.head()
```

```python
# Drop 'furnishingstatus' as the dummies have been created for it
df.drop(['furnishingstatus'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Check for duplicate rows
duplicate_rows = df.duplicated()

# Print the number of duplicate rows
print(duplicate_rows.sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
from sklearn.model_selection import train_test_split

# Specify this so that the train and test data set always have the same rows respectively
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# Rescaling the features by using Min-Max scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
# Check the correlation coefficients to see which variables are highly correlated
plt.figure(figsize = (18, 10))
sns.heatmap(df_train.corr(), annot = True, cmap = "YlGnBu")
plt.show()
```

```python
# Area seems correlated to price the most from the above
plt.figure(figsize = [6,6])
plt.scatter(df_train.area, df_train.price)
plt.show()
```

```python
# Split the dataset into training and testing sets for model evaluation
y_train = df_train.pop('price')
X_train = df_train
```

# **3. Regression Model Development**

```python
# Check all the columns of the dataframe
df.columns
```

```python
# Build a linear model
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.params
```

```python
print(lr_1.summary())
```

```python
# Check for the VIF values of the feature variables
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Drop highly correlated variables and insignificant variables
X = X_train.drop('semi-furnished', 1)
```

```python
# Build a second fitted model
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Print the summary of the model
print(lr_2.summary())
```

```python
# Calculate the VIFs again for the new model
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Drop highly correlated variables and insignificant variables
X = X.drop('bedrooms', 1)
```

```python
# Build a third fitted model
X_train_lm = sm.add_constant(X)

lr_3 = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Print the summary of the model
print(lr_3.summary())
```

```python
# Calculate the VIFs again for the new model
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Drop highly correlated variables and insignificant variables
X = X.drop('basement', 1)
```

```python
# Build a fourth fitted model
X_train_lm = sm.add_constant(X)

lr_4 = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Print the summary of the model
print(lr_4.summary())
```

```python
# Calculate the VIFs again for the new model
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Residual analysis of the train data
y_train_price = lr_4.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
plt.grid()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 16)                         # X-label
plt.show()
```

```python
# Applying the scaling on the test sets
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']
df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
df_test.describe()
```

```python
# Dividing into X_test and y_test
y_test = df_test.pop('price')
X_test = df_test
```

```python
# Adding constant variable to test dataframe
X_test_m4 = sm.add_constant(X_test)
```

```python
# Creating X_test_m4 dataframe by dropping variables from X_test_m4
X_test_m4 = X_test_m4.drop(["bedrooms", "semi-furnished", "basement"], axis = 1)
```

```python
# Making predictions using the fourth model
y_pred_m4 = lr_4.predict(X_test_m4)
```

# **4. Model Evaluation**

```python
# Plotting y_test and y_pred to understand the spread
fig = plt.figure()
plt.grid()
plt.scatter(y_test, y_pred_m4)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 16)                          # X-label
plt.ylabel('y_pred', fontsize = 16)   
plt.show()
```
